```python
# Group data
product_stock = df.groupby(['Product Name', 'Category'])['Stock Level'].mean().reset_index()

# Create figure
fig = go.Figure()

# Get unique categories
categories = product_stock['Category'].unique()

# Add one line per category
for cat in categories:
    data = product_stock[product_stock['Category'] == cat]
    
    fig.add_trace(go.Scatter(
        x=data['Product Name'],
        y=data['Stock Level'],
        mode='lines+markers',
        name=cat,  # 🔥 legend = category
        hovertemplate='<b>%{x}</b><br>Stock: %{y:.2f}<br>Category: ' + str(cat) + '<extra></extra>'
    ))

# Layout
fig.update_layout(
    title='📊 Product vs Average Stock Level (Category-wise)',
    xaxis_title='Product Name',
    yaxis_title='Average Stock Level',
    hovermode='x unified',
    template='plotly_white'
)

# Rotate product names (important)
fig.update_xaxes(tickangle=45)

# Show
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Step 1: Combine Category and Product Name
data['Category_Product'] = data['Category'] + " - " + data['Product Name']

# Step 2: Group by Category_Product to calculate average stock level
avg_stock = data.groupby('Category_Product').agg({
    'Stock Level': 'mean'
}).reset_index()

# Step 3: Create scatter plot
fig = px.scatter(
    avg_stock,
    x='Category_Product',
    y='Stock Level',
    text='Category_Product',  # shows product on hover
    title='Average Stock Level per Product',
    labels={'Stock Level':'Average Stock Level'},
    size_max=15
)

# Step 4: Improve readability
fig.update_traces(marker=dict(size=12, color='skyblue', line=dict(width=1, color='darkblue')))
fig.update_layout(xaxis_tickangle=-45, height=600)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Step 1: Group by Category and Product Name to calculate average Stock Level
avg_stock = data.groupby(['Category', 'Product Name']).agg({
    'Stock Level': 'mean'
}).reset_index()

# Step 2: Create scatter plot with swapped axes
fig = px.scatter(
    avg_stock,
    x='Stock Level',   # now x-axis is Average Stock Level
    y='Category',      # y-axis is Category
    color='Product Name',  # legend by Product Name
    title='Average Stock Level per Category by Product',
    labels={'Stock Level':'Average Stock Level', 'Category':'Category'},
    hover_data=['Product Name']
)

# Step 3: Improve readability
fig.update_layout(height=600)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Step 1: Ensure the correct column names
# Adjust if your dataset has slightly different names
stock_col = [c for c in data.columns if 'stock' in c.lower()][0]
pop_col = [c for c in data.columns if 'popularity' in c.lower()][0]

# Step 2: Group by Product Name to get average Stock Level and Popularity Index
avg_data = data.groupby('Product Name').agg({
    stock_col: 'mean',
    pop_col: 'mean'
}).reset_index()

# Step 3: Create scatter plot
fig = px.scatter(
    avg_data,
    x=stock_col,
    y=pop_col,
    color='Product Name',  # legend by Product Name
    title='Average Stock Level vs Popularity Index per Product',
    labels={stock_col:'Average Stock Level', pop_col:'Average Popularity'},
    hover_data=['Product Name']
)

# Step 4: Improve readability
fig.update_traces(marker=dict(size=12, line=dict(width=1, color='DarkSlateGrey')))
fig.update_layout(height=600)

fig.show()
```
